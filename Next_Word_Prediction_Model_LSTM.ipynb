```python
faqs='''
The United States, a vast and diverse country, stretches from the Atlantic to the Pacific, encompassing a wide range of landscapes, climates, and cultures.
From the snowy peaks of the Rocky Mountains to the sun-drenched beaches of California, the U.S. offers both natural beauty and modern urban landscapes.
Its cities, like New York, Los Angeles, and Chicago, are cultural hubs, attracting millions with their unique blend of history, art, and innovation.
The country has long been a global leader in technology, entertainment, and politics, and its history is shaped by waves of immigration, which have created a rich tapestry of traditions and values.
Despite its size and complexity, the United States continues to be a symbol of freedom and opportunity for many around the world.
'''
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer

#import class
```

```python
tokenizer=Tokenizer()

#create an object of tokenizer class
```

```python
tokenizer.fit_on_texts([faqs])

#tokenizer sends faqs text inside
#can send multiple text thats why we use text
```

```python
#unique word tokenization

#create a dictionary
tokenizer.word_index
```

```python
input_sequences=[]

#splitting all sentences on line changing basis
for sentence in faqs.split('\n'):
    tokenized_sentence=tokenizer.texts_to_sequences([sentence])[0]
    #send sentences and converts them into numbers and sequences

    for i in range(1,len(tokenized_sentence)):
      #create n gram model
      # The append method only takes the element as an argument.
      # Assign the slice to n_gram and then append it

      #i +1 means 1+1 means 2 as in next word predicted
      n_gram = tokenized_sentence[:i+1]
      input_sequences.append(n_gram)
```

```python
  input_sequences

  #print numerical representation of sequences
```

```python
#take every sentence length and add it to list
max_len=max([len(x) for x in input_sequences])
```

```python
#zero padding utility
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences=pad_sequences(input_sequences,maxlen=max_len,padding='pre')
```

```python
padded_input_sequences

#for every sentence first few are input 0... 1 and last is output 13
```

```python
X=padded_input_sequences[:,:-1]

#take all rows except last column
```

```python
y=padded_input_sequences[:,-1]

#take all last rows last column only
```

```python
len(tokenizer.word_index)
#total number of words
```

```python
#one hot encoding
#uses multi class classification

#converts word into probability
#always add +1 because unique word count starts from 1 and not 0

from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=len(tokenizer.word_index)+1)
```

```python
y.shape
```

```python
X,y
```

```python
#architecture
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
#build model
model=Sequential()

#len is total vocabulary
#input len is max words per sentence
model.add(Embedding(len(tokenizer.word_index)+1,100,input_length=max_len-1))
#150 nodes in LSTM
#working well

model.add(LSTM(150))
model.add(Dense(len(tokenizer.word_index)+1,activation='softmax'))

#dense layer requires as many nodes as vocabulary
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

#multi class classification so categorical cross entropy
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)

#use 100 epochs because
```

```python
#can change text but has to be within the unique word corpus to give good predictions
text="united states"

#run loop more to get more accurate predictions
for i in range(100):
  #tokenize

  token_text=tokenizer.texts_to_sequences([text])[0]
  #padding

  padded_token_text =pad_sequences([token_text],maxlen=max_len-1,padding='pre')


  #prediction
  #model.predict(padded_token_text)
  #model.predict(padded_token_text).shape
  import numpy as np

  pos=np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index==pos:
      text=text+ " "+ word
      print(text)
```
